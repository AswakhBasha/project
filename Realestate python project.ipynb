```python
  from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict([[789,299,661,552,612,990]])
```

```python
 from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.metrics import f1_score
from sklearn.preprocessing import KBinsDiscretizer

# Binning the Price into categories (e.g., 3 bins: low, medium, high)
kbin = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
y_train_binned = kbin.fit_transform(y_train.values.reshape(-1, 1)).astype(int)
y_test_binned = kbin.transform(y_test.values.reshape(-1, 1)).astype(int)
y_pred_binned = kbin.transform(y_pred.reshape(-1, 1)).astype(int)

# F1 score calculation
f1 = f1_score(y_test_binned, y_pred_binned, average='weighted')
print("F1 Score: ", f1)
```
